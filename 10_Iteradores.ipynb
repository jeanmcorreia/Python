```python
class ColecaoNumeros:
  def __init__(self, numero_max):
    self.max = numero_max
  def __iter__(self):
    self.numero_atual = 1
    return self

  def __next__(self):
    if self.numero_atual <= self.max:
      retorno = self.numero_atual
      self.numero_atual += 1
      return retorno
    else:
      raise StopIteration

colecao = ColecaoNumeros(6)
for item in colecao:
  print(item)
```

```python
def meu_range(num):
  local_num = 0
  while local_num < num:
    yield local_num
    local_num +=1

for i in meu_range(10):
  print(i)
```

```python
def anos():
  yield '2000'
  yield '2001'
  yield '2002'
  yield '2003'
  yield '2004'
  yield '2005'

for indice, valor in enumerate(anos()):
  print(indice, valor)
```

```python
def gen1():
  yield 1
  yield 2
  yield 3

def gen2():
  for i in gen1():
    yield i, 'a'
    yield i, 'b'
    yield i, 'c'


for x, y in gen2():
  print(x,y)
```
