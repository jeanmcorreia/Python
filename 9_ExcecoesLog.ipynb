```python
print("ínicio")
lista = [1,2,3]

try:
  print(lista[2])
except IndexError as error1:
  print(f"Erro de acesso ao índice {error1}")
except:
  print("Ocorreu outro erro")
else:
  print("Executa se não houver erro")

print("Fim")
```

```python
print("ínicio")
lista = [1,2,3]
numero = 0
try:
  divisao = 10 / numero
  print(lista[12])
except IndexError as error1:
  print(f"Erro de acesso ao índice {error1}")
except ZeroDivisionError as erro2:
  print(f"Erro de divisão por zero: {erro2}")
except Exception as erro3:
  print(f"Ocorreu outro erro: {erro3}")
else:
  print("Executa se não houver erro")

print("Fim")
```

```python
def printa_positivo(numero):
  if numero < 0:
    raise ValueError("Valor não pode ser negativo")
  print(numero)

try:
  printa_positivo(-1)
except ValueError as erro:
  print(f"O erro é: {erro}")
except Exception as erro1:
  print(f"Erro qualquer: {erro1}")
```

```python
def printa_positivo(numero):
  assert(numero >= 0)
  print(numero)

try:
  printa_positivo(-1)
except AssertionError as erro:
  print(f"O erro é: {erro}")
except Exception as erro1:
  print(f"Erro qualquer: {erro1}")
```
