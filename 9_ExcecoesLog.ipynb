```python
def printa_positivo(numero):
  if numero < 0:
    raise ValueError("Valor não pode ser negativo")
  print(numero)

try:
  printa_positivo(-1)
except ValueError as erro:
  print(f"O erro é: {erro}")
except Exception as erro1:
  print(f"Erro qualquer: {erro1}")
```

```python
def printa_positivo(numero):
  assert(numero >= 0)
  print(numero)

try:
  printa_positivo(-1)
except AssertionError as erro:
  print(f"O erro é: {erro}")
except Exception as erro1:
  print(f"Erro qualquer: {erro1}")
```

```python
# Crie uma função que receba duas strings que serão convertidas em numeros
# para serem somadas. se ao realizar o casting ocorrer um erro, gere uma exceção
# informando o motivo

def soma_string(str1, str2):
  try:
    num1, num2 = int(str1), int(str2)
  except ValueError as erro:
    print(f"Não foi possível fazer casting: {erro}")
  else:
    return num1 + num2
  finally:
    print("Fim do programa")

soma_string('joao', '20')

# Solução

def soma(str1, str2):
  try:
    num1 = float(str1)
    num2 = float(str2)
    return num1 + num2
  except:
    raise Exception("Falha no casting")
```

```python
# 4- Crie uma classe que represente um caracter (string de tamanho 1), use
# propriedades ou crie uma função para isso (mas deixe valor privado) e caso
# o usuário tente inserir um texto gere uma exceção dizendo o motivo.

class Caracter:
  def __init__(self, caracter):
    self.__caracter = ''
    self.caracter = caracter

  @property
  def caracter(self):
    return self.__caracter

  @caracter.setter
  def caracter(self, value):
    if len(value) > 1:
      raise Exception("Caracter deve ter no máximo tamanho 1")

    self.__caracter = value

letra = Caracter("a")
print(letra.caracter)

try:
  letra.caracter = 'b'
except Exception as ex:
  print(str(ex))

print(letra.caracter)
```
